## Data cleaning

```python
df.columns
```

```python
main_columbs = ['Time',
    'Day_of_week',
    'Age_band_of_driver',
    'Sex_of_driver',
    'Type_of_vehicle',
    'Vehicle_movement',
    'Driving_experience',
    'Service_year_of_vehicle',
    'Cause_of_accident',
    'Accident_severity'
]
df1 = df[main_columbs]
df1.columns
```

```python
df1
```

```python
null_number= df1.isnull().sum()
```

```python
df1['Type_of_vehicle']=df1['Type_of_vehicle'].fillna('unknown')
df1['Type_of_vehicle']
```

```python
df1['Type_of_vehicle'] = df1['Type_of_vehicle'].replace('Public (> 45 seats)','Public')
df1['Type_of_vehicle'] = df1['Type_of_vehicle'].replace('Lorry (41?100Q)','Lorry')
df1['Type_of_vehicle'] = df1['Type_of_vehicle'].replace('Lorry (11?40Q)','Lorry')
df1['Type_of_vehicle']
```

```python
df_mode = df1['Vehicle_movement'].mode
df1['Vehicle_movement'] = df1['Vehicle_movement'].fillna('df_mode')
```

```python
df1['Driving_experience'] =df1['Driving_experience'].replace('Above 10yr','10-15yr')
```

```python
driver_experience_mod = df1['Driving_experience'].mode
df1['Driving_experience']= df1['Driving_experience'].fillna(driver_experience_mod)
```

```python
service_year_mod = df1['Service_year_of_vehicle'].mode
df1['Service_year_of_vehicle'] = df1['Service_year_of_vehicle'].fillna(service_year_mod)
```

# Line Chart

## Descriptive Questions and Answers
1. How does accident severity differ by time of day?
2. Are there differences in accident severity based on the day of the week?

```python
#1. How does accident severity differ by time of day?
df['Hour'] = pd.to_datetime(df['Time'], format='%H:%M').dt.hour
sns.boxplot(x='Accident_severity', y='Hour', data=df, palette='pastel')
plt.title('Accident Severity by Time of Day',fontweight = 'bold', fontsize = 15)
plt.xlabel('Accident Severity',fontweight = 'bold', fontsize = 15)
plt.ylabel('Hour of Day',fontweight = 'bold', fontsize = 15)
plt.show()
```

```python
#2. Are there differences in accident severity based on the day of the week?
severity_counts = df.groupby('Day_of_week')['Accident_severity'].value_counts().unstack().fillna(0)
severity_counts.plot(kind='bar', stacked=True)
plt.title('Accident Severity by Day of Week',fontweight = 'bold', fontsize = 20)
plt.xlabel('Day of Week',fontweight = 'bold', fontsize = 15)
plt.ylabel('Count',fontweight = 'bold', fontsize = 15)

```
